```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

#function format to currency
def cformat(x):
    return "${:,.2f}".format(x)
```

# Player Count

```python
#Total Players : remove duplicates then count

totalplayer = purchase_data['SN'].unique()
totalplayer_df = pd.Series(totalplayer)
total_players = pd.DataFrame([{"Total Players": totalplayer_df.count()}])
total_players

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items: remove duplicates

unique = purchase_data['Item ID'].unique()
uniqueitems = len(unique)

#total revenue: summation
totalrevenue = purchase_data["Price"].sum()
total_revenue = "${:,.2f}".format(totalrevenue)

#no of purchase: count
numberofpurchases = purchase_data["Price"].count()

#average price: mean
averageprice = purchase_data["Price"].mean()
average_price = "${:,.2f}".format(averageprice)

#create dataframe
second_list = pd.DataFrame([{"Number of Unique Items": uniqueitems,"Average Price":average_price,
               "Number of Purchases":numberofpurchases,"Total Revenue": total_revenue}])
second_list[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics: create new data frame without duplicates
player_gender = purchase_data[["SN", "Gender"]]
player_gender_df = pd.DataFrame(player_gender)
no_duplicates = player_gender_df.drop_duplicates()

#Count ot total players per gender
count_male = (no_duplicates['Gender']=='Male').sum()
count_female = (no_duplicates['Gender']=='Female').sum()
count_others = (no_duplicates['Gender']=='Other / Non-Disclosed').sum()
totalcount = count_male + count_female + count_others

#Percentage of players per gender
percentmale = (count_male/totalcount)*100
percentfemale = (count_female/totalcount)*100
percentothers = (count_others/totalcount)*100


#DataFrame for Gender Demographics
Gender_Player_Count = pd.DataFrame({"": ["Male", "Female", "Other/ Non-Disclosed"],"Total Count": [count_male, count_female, count_others],
                                 "Percentage of Players":[percentmale, percentfemale, percentothers]})
Gender_Player_Count["Percentage of Players"] = Gender_Player_Count["Percentage of Players"].map("{:.2f}".format)
Gender_Player_Count_indexed = Gender_Player_Count.set_index("")
Gender_Player_Count_indexed
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Gender Counts
Male = (purchase_data['Gender']=='Male').sum()
Female = (purchase_data['Gender']=='Female').sum()
Others = (purchase_data['Gender']=='Other / Non-Disclosed').sum()
total = Male + Female + Others
groupofmales = purchase_data.loc[purchase_data["Gender"] == "Male", :]

#MALE
tot_purch_male = groupofmales["Price"].sum()
ave_purch_male = tot_purch_male / Male
avg_person_male = tot_purch_male / count_male

groupoffemales = purchase_data.loc[purchase_data["Gender"] == "Female", :]
#FEMALE
tot_purch_female = groupoffemales["Price"].sum()
ave_purch_female = tot_purch_female / Female
avg_person_female = tot_purch_female / count_female


groupofothers = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
#OTHERS
tot_purch_others = groupofothers["Price"].sum()
ave_purch_others = tot_purch_others / Others
avg_person_others = tot_purch_others/ count_others


Gender_Purchase = pd.DataFrame({"Gender": ["Female", "Male", "Other/ Non-Disclosed"],"Purchase Count": [Female, Male, Others],
                                 "Average Purchase Price":[ave_purch_female, ave_purch_male, ave_purch_others],
                                 "Total Purchase Value":[tot_purch_female,tot_purch_male,tot_purch_others], 
                                 "Avg Total Purchase per Person": [avg_person_female, avg_person_male, avg_person_others]})
gender_purchase_table = Gender_Purchase.set_index("Gender")
gender_purchase_table["Average Purchase Price"] = gender_purchase_table["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_table["Total Purchase Value"] = gender_purchase_table["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_table["Avg Total Purchase per Person"] = gender_purchase_table["Avg Total Purchase per Person"].map("${:,.2f}".format)
gender_purchase_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
newframe = purchase_data[["Age", "SN"]]
newframe_df = pd.DataFrame(newframe)
newframe_df = newframe.drop_duplicates()
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,99.9]
names = ["<10","10-14","15-19","20-24","25-29", "30-34","35-39","40+"]
newframe_df["Total Count"] = pd.cut(newframe_df["Age"], bins, labels=names)
new_data_1 = newframe_df["Total Count"].value_counts()
newframe_data_2 = pd.DataFrame(new_data_1)
newframe_data_2["Percentage of Players"] = [x for x in new_data_1 / (totalplayer_df.count()) *100]
newframe_data_3 = newframe_data_2.sort_index()
newframe_data_3["Percentage of Players"] = newframe_data_3["Percentage of Players"].map("{:.2f}".format)
newframe_data_3
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_analysis_1 = purchase_data[["Age", "Price"]]
purchase_analysis_2 = pd.DataFrame(purchase_analysis_1)

#Average Purchase Price
purchase_analysis_2[""] = pd.cut(purchase_analysis_2["Age"], bins, labels=names)
purchase_analysis_a = purchase_analysis_2.loc[purchase_analysis_2[""] == "<10"]
count_a = purchase_analysis_a["Age"].count()
purchase_analysis_a_ave = purchase_analysis_a["Price"].mean()
purchase_analysis_a_tot = purchase_analysis_a["Price"].sum()

purchase_analysis_b = purchase_analysis_2.loc[purchase_analysis_2[""] == "10-14"]
purchase_analysis_b_ave = purchase_analysis_b["Price"].mean()
purchase_analysis_b_tot = purchase_analysis_b["Price"].sum()

purchase_analysis_c = purchase_analysis_2.loc[purchase_analysis_2[""] == "15-19"]
purchase_analysis_c_ave = purchase_analysis_c["Price"].mean()
purchase_analysis_c_tot = purchase_analysis_c["Price"].sum()

purchase_analysis_d = purchase_analysis_2.loc[purchase_analysis_2[""] == "20-24"]
purchase_analysis_d_ave = purchase_analysis_d["Price"].mean()
purchase_analysis_d_tot = purchase_analysis_d["Price"].sum()

purchase_analysis_e = purchase_analysis_2.loc[purchase_analysis_2[""] == "25-29"]
purchase_analysis_e_ave = purchase_analysis_e["Price"].mean()
purchase_analysis_e_tot = purchase_analysis_e["Price"].sum()

purchase_analysis_f = purchase_analysis_2.loc[purchase_analysis_2[""] == "30-34"]
purchase_analysis_f_ave = purchase_analysis_f["Price"].mean()
purchase_analysis_f_tot = purchase_analysis_f["Price"].sum()

purchase_analysis_g = purchase_analysis_2.loc[purchase_analysis_2[""] == "35-39"]
purchase_analysis_g_ave = purchase_analysis_g["Price"].mean()
purchase_analysis_g_tot = purchase_analysis_g["Price"].sum()

purchase_analysis_h = purchase_analysis_2.loc[purchase_analysis_2[""] == "40+"]
purchase_analysis_h_ave = purchase_analysis_h["Price"].mean()
purchase_analysis_h_tot = purchase_analysis_h["Price"].sum()


purchase_analysis_3 = pd.DataFrame(purchase_analysis_2[""].value_counts())
purchase_analysis_3["Average Purchase Price"] = [purchase_analysis_d_ave,purchase_analysis_c_ave,
                                                purchase_analysis_e_ave, purchase_analysis_f_ave,
                                                purchase_analysis_g_ave, purchase_analysis_b_ave,
                                                purchase_analysis_a_ave, purchase_analysis_h_ave]
purchase_analysis_3["Total Purchase Value"] = [purchase_analysis_d_tot,purchase_analysis_c_tot,
                                                purchase_analysis_e_tot, purchase_analysis_f_tot,
                                                purchase_analysis_g_tot, purchase_analysis_b_tot,
                                                purchase_analysis_a_tot, purchase_analysis_h_tot]
purchase_analysis_4 = purchase_analysis_3.sort_index()
purchase_analysis_4["Avg Total Purchase per Person"] = [p for p in purchase_analysis_4["Total Purchase Value"] / newframe_data_3["Total Count"]]
purchase_analysis_5 = purchase_analysis_4.rename(columns={"":"Purchase Count"})
purchase_analysis_5["Average Purchase Price"] = purchase_analysis_5["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_5["Total Purchase Value"] = purchase_analysis_5["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_5["Avg Total Purchase per Person"] = purchase_analysis_5["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_5
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topspenders =  purchase_data[["SN","Price"]]
top_grouped = topspenders.groupby("SN")
top_agg = pd.DataFrame(top_grouped["Price"].agg(np.sum))
top_agg.reset_index(level=0, inplace=True)

listplayers=pd.DataFrame(purchase_data["SN"].value_counts())
listplayers.reset_index(level=0, inplace=True)
listplayers
listplayers_renamed = listplayers.rename(columns={"index":"SN","SN":"Purchase Count"})
merged_players_totpurch = pd.merge(top_agg, listplayers_renamed, on="SN", how="outer")
merged_players_totpurch["Average Purchase Price"] = [a for a in merged_players_totpurch["Price"] / merged_players_totpurch["Purchase Count"]]
merged_sorted = merged_players_totpurch.sort_values("Price", ascending=False)
merged_organize = merged_sorted[["SN","Purchase Count", "Average Purchase Price", "Price"]]
merged_organize_index = merged_organize.set_index("SN")
merged_organize_renamed = merged_organize_index.rename(columns={"Price":"Total Purchase Value"})
merged_organize_renamed["Average Purchase Price"] = merged_organize_renamed["Average Purchase Price"].apply(cformat)
merged_organize_renamed["Total Purchase Value"] = merged_organize_renamed["Total Purchase Value"].apply(cformat)
final_topspenders = merged_organize_renamed.head()
final_topspenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item Id, Item Name, Total Purchase Value
popitem =  purchase_data[["Item ID", "Item Name","Price"]]
popitem.drop_duplicates(subset="Item Name", keep='first')
popitem_grouped = popitem.groupby(["Item ID","Item Name"], as_index=False)
popitem_grouped= pd.DataFrame(popitem_grouped["Price"].agg(np.sum))
popitem_renamed = popitem_grouped.rename(columns={"Price":"Total Purchase Value"})
popitem_droped = popitem_renamed.drop_duplicates(subset="Item Name", keep='first')

#Count and Price
count_ = pd.DataFrame(purchase_data["Item Name"].value_counts())
count_.reset_index(level=0, inplace=True)
count_renamed = count_.rename(columns={"Item Name": "Purchase Count","index":"Item Name"})
item_price = purchase_data[["Item Name","Price"]]
item_nodup = item_price.drop_duplicates()
count_price = pd.merge(count_renamed, item_nodup, on="Item Name")
count_price_rename = count_price.rename(columns={"Price":"Item Price"})
count_price_droped = count_price_rename.drop_duplicates(subset="Item Name", keep='first')


#Merge Tables: Setting Indices, Changing Format
merged_popitems = pd.merge(count_price_droped, popitem_droped, on="Item Name")
merged_popitems_sorted = merged_popitems.sort_values("Purchase Count", ascending=False)
merged_indexed = merged_popitems_sorted.set_index(["Item ID", "Item Name"])
merged_indexed["Item Price"] =  merged_indexed["Item Price"].apply(cformat)
merged_indexed["Total Purchase Value"] = merged_indexed["Total Purchase Value"].apply(cformat)
merged_indexed.head(10)

```
